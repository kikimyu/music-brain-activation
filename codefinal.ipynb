#### I: Locate MRI data automatically based on file extension/name
This code locates two kinds of MRI data files: 
1. High resolution MRI scan of the anatomy of the human brain called the **anatomical** file 
2. Statistical map called the **functional** file, displaying which regions are consistently active throughout studies related to the term 'music'

It does so using the **load_img** module from the **nilearn** package. The **glob** package is used to automatically locate the two kinds of files.

```python
import glob #Import glob to find files in directory
from nilearn.image import load_img #Import image loading module to load data files as images

anatomical_img = load_img(glob.glob("anatomical.nii")) #Loads anatomical image
functional_img = load_img(glob.glob("*0.01.nii")) #Loads functional image
```

#### II: Visualize the statistical map on top of the anatomical scan
This code plots the data from the functional file onto the image of the brain anatomy.

It does so using the **plotting** module from **nilearn**, which takes the functional data and layers it onto the anatomical image. 

Voxel **threshold** is set to 0.1, as to exclude any voxels below this value. 

```python
from nilearn import plotting  #Import the plotting module to visualize brain data on top of image

#Plots functional data onto anatomical image
plotting.plot_stat_map(functional_img, bg_img=anatomical_img,
                       cut_coords=None, #Set cut coordinations to None to automatically find the best coordinates
                       title="Music Related Activation",
                       threshold=0.1, cmap="PiYG")

plotting.show() #Shows plot
```

#### III: Plot a histogram of the values from the functional data
Using **matplotlib**, this code takes the data from the functional file and adds its values into a histogram.

Note that only the positive values are being used!

```python
###III: Plot histogram of values from functional data file

import matplotlib.pyplot as plt #Import plt library to plot histogram

data = functional_img.get_fdata() #Retrieves data from functional image
positive_data = data[data > 0] #Selects all positive values

fig1 = plt.figure(figsize=(10,6)) #Creates empty figure

#Adds axis titles
plt.title(" Functional MRI Music Activation Data ")
plt.xlabel("Voxel Intensity")
plt.ylabel("Voxel Frequency")

plt.hist(positive_data, color="lavenderblush", 
         edgecolor="black") #Adds edges to bar for clarity
plt.show()
```
